```python
# ============================================================
# EDIT THESE VARIABLES to match your setup
# ============================================================

# Git repo URL
REPO_URL = "https://github.com/fsotosan/surveillance"

# Where the repo lives in your Drive. If it doesn't exist yet,
# it will be cloned here.
PARENT_DIR = "/content/drive/MyDrive/dev/surveillance"

# Directory name inside PARENT_DIR that the repo is/will be cloned to
REPO_DIR = "surveillance"

# Path to your training images directory (onboarded frame images)
# Can be absolute or relative to {PARENT_DIR}/{REPO_DIR}
TRAINING_IMAGES = "../images"

# Path to your training database (training.db from a prior `onboard` run)
# Can be absolute or relative to {PARENT_DIR}/{REPO_DIR}
TRAINING_DB = "../db/training.db"

# Training hyperparameters
EPOCHS = 500
IMGSZ = 640
BATCH = 16

# Set to True to export the dataset and generate labeled previews
# without actually training.
EXPORT_ONLY = False
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os
from pathlib import Path

REPO_PATH = Path(PARENT_DIR) / REPO_DIR

if REPO_PATH.exists():
    print(f"Updating existing repo at {REPO_PATH}")
    %cd $REPO_PATH
    !git pull
else:
    print(f"Cloning repo into {REPO_PATH}")
    Path(PARENT_DIR).mkdir(parents=True, exist_ok=True)
    %cd $PARENT_DIR
    !git clone $REPO_URL

%cd $REPO_PATH
```

```python
%env SURVEILLANCE_TRAINING_IMAGES=$TRAINING_IMAGES
%env SURVEILLANCE_TRAINING_DB=$TRAINING_DB

flags = f"--epochs {EPOCHS} --imgsz {IMGSZ} --batch {BATCH}"
if EXPORT_ONLY:
    flags += " --export-only"

!python -m surveillance train $flags
```
